<h4> Performing operations on CSV/Excel file

```python
import pandas as pd
```

```python
# Reading a Excel file and head func
data = pd.read_excel('my_data.xlsx')
data.head(2)
```

```python
# To check if any columns has null values, .sum() gives the sum of null values
data.isnull().sum()
```

```python
#data.dropna(inplace=True)  #deletes the values in existing dataset
```

```python
# Drop Null values
data.dropna()
```

```python
# Fill the Null values with 100
data.fillna(100)
```

```python
# To Access a columns
print(data['Pulse'])
```

```python
# To fill the Null values of a one column with a number
data['Calories'].fillna(100)
```

```python
#To replace the Null values of Calories with mean
x=data['Calories'].mean()
print(x)
data['Calories'].fillna(x)
```

```python
#To replace the Null values of Calories with Mode
y=data['Calories'].mode()[0]
print(y)
data['Calories'].fillna(y)
```

```python
#To replace the Null values of Calories with Median
z=data['Calories'].median()
print(z)
data['Calories'].fillna(z)
```

```python
# Drop the Null values of Date Column
data.dropna(subset=['Date'])
```

```python
# To handle the outliers and hange its values
data.loc[7,'Duration']=45
data
```

```python
# To make a columns in range
for i in data.index:
    if data.loc[i,'Pulse']>110:
        data.loc[i,'Pulse']=110
data
```

```python
data.duplicated()
```

```python
data.drop_duplicates()
```

```python
data.corr()
```

```python
data.sample()
```

<h1> DataFrame <h1>

```python
import pandas as pd
```

```python
# Creating a Data Frame Using a Dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [24, 27, 22, 32, 29],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'Salary': [70000, 80000, 65000, 90000, 75000]
}
# Create DataFrame
df = pd.DataFrame(data)
df
```

```python
# Accessing rows and columns
print(df['Name'])     # Name column
print(df.loc[0])      # First row
print(df.loc[0:2])    # First two rows
print(df.loc[[0,2]])  # First and third row   OR print(df.loc[0:2:2])
```

```python
# Add a New Column
df['Department'] = ['HR', 'IT', 'Finance', 'HR', 'IT']
df
```

```python
# Filter Rows
filtered_df = df[df['Age'] > 25]
filtered_df
```

```python
# Delete a Column
del df['Department']
df
```

```python
# Delete a Row
df.drop(2, axis=0, inplace=True)
df
```

```python
# Modify a element in a row
df.loc[0,'Age']=25
df
```

```python
# Sort the dataframe
df.sort_values(by='Salary', ascending=False)
```

```python
# Rename the Columns
df.rename(columns={'Name':'First Name'}, inplace=True)
df
```

```python
# Add a row
new_row = {'First Name': 'Frank', 'Age': 30, 'City': 'San Francisco', 'Salary': 85000}
df = df._append(new_row, ignore_index=True)
df
```

```python
# get() function
df.get('Age')
df.get(['Age','Salary'])
```

```python
# Creating a DataFrame using List of Lists and Defining Column names
ls= [["Sara",19,"CSE"],["Bea",20,"Fashion"],["Ben",19,"CSE"]]
df1= pd.DataFrame(ls, columns=["Name","Age","Dept"])
df1
```

```python
# Creating a DataFrame using dictionary of Series
x={"Name": pd.Series(["g","e","k","o"]), "Age": pd.Series([1,2,3,4])}
df2= pd.DataFrame(x)
df2
```

```python
# Add a new column
df2["dept"]=pd.Series(["CSE","ECE","EEE","ME"])
df2
```

<h2> Converting a Dataset into DataFrame

```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
# To show all the rows of dataset
pd.options.display.max_rows=9999
```

```python
iris= load_iris()
df3= pd.DataFrame(iris.data, columns= iris.feature_names)
df3
```
