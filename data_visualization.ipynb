### Bar chart
It is used to create different categories of data

```python
age=[23,34,25,32,34,37]
salary=[15000,34000,37000,43000,29500,40000]
plt.bar(age,salary, color="#E9967A",edgecolor="black",label="Salary")
# Adding title to the plot
plt.title("salaries corresponds to age")
# Adding label to the X-axis
plt.xlabel("Age",fontsize=15)
# Adding label to the Y-axis
plt.ylabel("salary",fontsize=15)
# Displaying the legend
plt.legend()
# Showing the final plot
plt.show()
```

### Line chart
it is used to represent the trend line.

```python
# Plotting the salary data against age
plt.plot(age,salary, color="maroon",label="Salary",marker='o',lw="2",markersize=5)
# marker= x,X,+,1,2,3,*,<,>
# lw= line width
# Adding title to the plot
plt.title("salaries corresponds to age")
# Adding label to the X-axis
plt.xlabel("Age",fontsize=15)
# Adding label to the Y-axis
plt.ylabel("salary",fontsize=15)
# Displaying the legend
plt.legend()
# Showing the final plot
plt.show()
```

### Histogram 
use dto represent frequency distribution

```python
# Importing necessary libraries
import matplotlib.pyplot as plt
# Data for the histogram
data = [1, 1, 2, 3, 4, 4, 4, 5, 5, 6, 7, 8, 8, 9]

# Create the histogram
plt.hist(data, color="lightgreen", edgecolor="black")
# Adding label to the X-axis
plt.xlabel('Values')
# Adding label to the Y-axis
plt.ylabel('Frequency')
# Adding title to the plot
plt.title('Histogram')
# Showing the final plot
plt.show()
```

```python
# Importing necessary libraries
import matplotlib.pyplot as plt
# Data for the histogram
data = [1, 1, 2, 3, 4, 4, 4, 5, 5, 6, 7, 8, 8, 9]

# Create the histogram
plt.hist(data, bins=[1,2,3,4,5,6], color="lightgreen", edgecolor="black")
# Adding label to the X-axis
plt.xlabel('Values')
# Adding label to the Y-axis
plt.ylabel('Frequency')
# Adding title to the plot
plt.title('Histogram')
# Showing the final plot
plt.show()
```

### Box plot
used to represent 5 point summary <br>
q1 : ist Quartile= 25 percentile<br>
Q2 - 2nd quartile= 50 percent = median<br>
Q3 3rd quartile = 75 percentile<br>
min Q1-1.5*IQR<br>
max Q3+1.5*IQR<br>

Iqr= Inter Quartile Range = Q3-Q1<br>
Outliner = Data points>max or < min<br>


```python
# Importing necessary libraries
import matplotlib.pyplot as plt
data = [10, 15, 20, 25, 30, 35, 40,34,56,23,43,12,56,43,32,38,39,26,22.37]
plt.boxplot(data)
# Adding title to the plot
plt.title('Box Plot')
plt.grid()
# Showing the final plot
plt.show()
```
